```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r'C:\SACHIN RAWAL FILES\LAPTOP\PYTHON COURCE\DATA_VISUALIZATION\MATPLOTLIB\netflix_dataset.csv')
# print(df)
```

```python
df.dropna(subset=['show_id','type','title','director','cast','country','date_added','release_year','rating','duration','listed_in','description'])
type_count = df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_count.index,type_count.values,color =['lightblue','lightgreen'])
plt.title("Number of TV show and Movies")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Movies_vs_TV_show.png")
plt.show()
```

```python
rating_count = df['rating'].value_counts()
fig , ax = plt.subplots(1,2,figsize = (20,10))
ax[0].pie(rating_count,labels=rating_count.index,autopct="%1.1f%%",startangle=90)

ax[1].barh(rating_count.index,rating_count,color = 'green')
fig.suptitle("Percentage of Content Rating",fontsize = 20)
plt.savefig("Percetile_of_rating.png")
plt.tight_layout()
plt.show()
```

```python
movie_df = df[df['type']=='Movie'].copy()
movie_df['duration_int'] = movie_df['duration'].str.replace('min','').astype(int)

plt.figure(figsize=(6,8))
plt.hist(movie_df['duration_int'],bins=30,color='skyblue',edgecolor = 'purple')
plt.title("Distribution of Movie")
plt.xlabel("Duration(Minutes)")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.savefig("Distribution_of_Movie.png")
plt.show()
```

```python
release_count = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,7))
plt.scatter(release_count.index,release_count.values,color = 'green')
plt.plot(release_count.index,release_count.values,color = 'grey')
plt.title("Release Year Vs Number of Shows")
plt.xlabel("Release Year")
plt.ylabel("Number of shows")
plt.tight_layout()
plt.savefig("Release_Year_Vs_Number_of_Shows.png")
plt.show()
```

```python
country_count = df['country'].value_counts().head(13)
plt.figure(figsize=(10,8))
plt.barh(country_count.index,country_count.values,color = 'teal')
plt.title("Top 10 country by number of shows")
plt.xlabel("Number of Shows")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig('Country_release_show.png')
plt.show()
```

```python
contenrt_of_year = df.groupby(['release_year','type']).size().unstack().fillna(0)
#for tv show and movies
fig, ax = plt.subplots(1,2,figsize=(14,6))
ax[0].plot(contenrt_of_year.index, contenrt_of_year['Movie'], color='blue',label = 'data of movies')
# ax[0].set_title("Movies Release per year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("No of Movies")

ax[0].plot(contenrt_of_year.index,contenrt_of_year['TV Show'],color = 'orange',label = 'Data of Tv shows')
# ax[0].set_title("Tv Show Release per year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("No of TV Shows")
fig.legend(loc = 'upper left')
fig.suptitle("Data of Number of Shows release in year")


```
